```python
from collections import defaultdict as Dictionary
import pandas as pd
```

```python
hydro = 'electricity_bill'
water = 'water_bill'
utlts = 'utility_bill'
ul = 'Upper_Limit'
ll = 'Lower_Limit'
weight = 'weightage'

data_dict = Dictionary()
```

```python
data_dict[hydro] = {weight:10,ll:50, ul:1000}
data_dict[water] = {weight:5, ll:50, ul:300}
data_dict[utlts] = {weight:10, ll:30, ul:200}
```

```python
sustainable_score = 0
```

```python
def score_calc(current_values, sustainable_score = 0):
  for category in data_dict.keys():
    category_based_score = (data_dict.get(category).get(weight)*sustainable_score)/100
    sustainable_score -= category_based_score
    sustainable_score += (category_based_score + (data_dict.get(category).get(weight)*(current_values.get(category) - data_dict.get(category).get(ll)))/(data_dict.get(category).get(ul) - data_dict.get(category).get(ll)))/2
  return sustainable_score
```

```python
user_data = pd.read_csv('user_data.csv')
user_data.head(10)
```

```python
user_data.iloc[:,3:]
```

```python
def fn(x):
  current_values = {hydro:x[0], water:x[1], utlts:x[2]}
  return score_calc(current_values, sustainable_score)
```

```python
user_data["sustainable_score"] = user_data.iloc[:,3:].apply(fn, axis=1)
```
